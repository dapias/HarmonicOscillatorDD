## Simulation of a Harmonic Oscillator in the microcanonical ensemble with the Verlet algorithm

```julia
include("../src/MicrocanonicalSimulation.jl")
```

```julia
N = 1
rho = 1.0
dt = 0.00005
T = 1.0
runtime = 10.0;
```

```julia
time, energy, kinetic, potential, temperature, atoms = MicrocanonicalSimulation.run(runtime, rho, dt, T, N);
```

```julia
using PyPlot
```

```julia
plot(time, kinetic, "b", time, potential, "r", time, energy, "g");
```
